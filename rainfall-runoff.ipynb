## Explore data

```python
import pandas as pd

%matplotlib inline
```

Data from https://nrfa.ceh.ac.uk/data/search 

* download the the 2 files for catchment 3001: `3001_cdr.csv` and `3001_gdf.csv`
* upload to Watson Studio from menu on right side of notebook
* click on file and select import as pandas DataFrame

```python
rain = pd.read_csv('data/3001_cdr.csv',skiprows=16)
rain.columns = ['date','rainfall','x']
rain = rain.drop('x', axis=1)
rain = rain.set_index('date')

flow = pd.read_csv('data/3001_gdf.csv',skiprows=16)
flow.columns = ['date','flow','x']
flow = flow.drop('x', axis=1)
flow = flow.set_index('date')


```

```python
rain.head()
```

```python
flow.head()
```

```python
merged = rain.merge(flow, left_on='date', right_on='date')
merged.head()
```

```python
merged.plot(figsize=(14,5));
```

Anything you notice here from this plot? Shape, time delays, how is runoff related to rainfall?

```python
merged.plot.scatter(x='rainfall',y='flow')
```
